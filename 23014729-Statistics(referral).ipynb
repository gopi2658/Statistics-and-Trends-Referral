# Applied Data Science

## Statistics and Trends

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def analyze_and_plot(df):
    # Selecting numeric columns for analysis
    numeric_cols = df.select_dtypes(include=[np.number])

    # Basic descriptive statistics
    desc = numeric_cols.describe()
    print("Descriptive Statistics:\n", desc)

    # Mean and median
    mean_median = numeric_cols.agg(['mean', 'median'])
    print("\nMean and Median:\n", mean_median)

    # Standard deviation
    std_dev = numeric_cols.std()
    print("\nStandard Deviation:\n", std_dev)

    # Skewness and kurtosis
    skewness = numeric_cols.skew()
    kurtosis = numeric_cols.kurt()
    print("\nSkewness:\n", skewness)
    print("\nKurtosis:\n", kurtosis)

    # Correlation matrix
    corr_matrix = numeric_cols.corr()
    print("\nCorrelation Matrix:\n", corr_matrix)

    # Plotting the heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title('Correlation Matrix Heatmap')
    plt.show()

# Load the dataset
file_path_new = './train.csv'
df = pd.read_csv(file_path_new)

# Call the function to analyze and plot
analyze_and_plot(df)
```

```python
# Histogram Plot of Passenger Class Distribution
def plot_pclass_distribution(df):
    df['Pclass'].value_counts().plot(kind='bar')
    plt.title('Passenger Class Distribution')
    plt.xlabel('Class')
    plt.ylabel('Count')
    plt.show()

plot_pclass_distribution(df)
```

```python
# Scatter Plot of Fare vs Age
def plot_fare_age_relation(df):
    plt.scatter(df['Age'], df['Fare'])
    plt.title('Fare vs Age')
    plt.xlabel('Age')
    plt.ylabel('Fare')
    plt.show()

plot_fare_age_relation(df)
```

```python
# Boxplot grouped by Pclass Fare Distribution by Class
def plot_fare_distribution(df):
    df.boxplot(column='Fare', by='Pclass')
    plt.title('Fare Distribution by Class')
    plt.xlabel('Class')
    plt.ylabel('Fare')
    plt.show()

plot_fare_distribution(df)
```
